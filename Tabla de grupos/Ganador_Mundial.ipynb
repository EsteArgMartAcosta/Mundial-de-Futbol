```python
import pandas as pd
from string import ascii_uppercase as alfabeto
import pickle
```

**Leer la página y sus tablas**

```python
todas_tablas = pd.read_html('https://en.wikipedia.org/wiki/2022_FIFA_World_Cup')
```

```python

print("Hay todas estas tablas en la página",len(todas_tablas))

# Ver las primeras filas de cada tabla para identificar cuál es la que quieres
for i, tabla in enumerate(todas_tablas):
    print(f"Tabla {i}:")
    print(tabla.head(), "\n")
```

```python
print(todas_tablas[18])
```

```python
alfabeto
```

```python
dict_tablas = {}
for letra ,i in zip(alfabeto,  range(18,74, 7)):
  df = todas_tablas[i]
  df.rename(columns = {df.columns[1]: 'Team'}, inplace= True)
  df.pop('Qualification')
  dict_tablas[f'Grupo {letra}'] = df
```

```python
dict_tablas.keys()
```

```python
dict_tablas['Grupo A']
```

```python
with open('dict_table', 'wb') as output:
  pickle.dump(dict_tablas, output)
```
